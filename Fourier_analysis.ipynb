```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 10, 200) # 0 to 10 with 200 points.  Treat as time in days
```

```python
# Want amplitude of 0.5 and frequency of 2 cycles/day
freq = 2
s = 0.5*np.sin(freq*2*np.pi*t)
plt.plot(t, s)
```

```python
# Want amplitude of 0.5 and frequency of 2 cycles/day
c = 0.5*np.cos(freq*2*np.pi*t)
plt.figure()
plt.plot(t, c)
plt.figure()
plt.plot(t, s)
plt.plot(t, c)
```

```python
# Correlation
#dist = np.sqrt(np.sum((s-c)**2))
corr_matrix = np.corrcoef(s, c)
# Approx zero
corr_matrix
print("correlation is ", corr_matrix[0,1])
```
