```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([-4, -2, 2, 4, -4])
y = np.array([0, 2, 2, 0, 0])
plt.figure()
plt.plot(x, y, lw=2)
plt.axis('square')
```

```python
def rotateByDegrees(deg):
    mat = np.zeros([3,3])
    angle = deg*np.pi/180.0
    mat[0,0] = np.cos(angle)
    mat[0,1] = np.sin(angle)
    mat[1,0] = -np.sin(angle)
    mat[1,1] = np.cos(angle)
    mat[2,2] = 1
    return mat
    

# Shift by 4 to the right.
# note that np.eye(x) would fill the identity parts.
right4 = np.array([[1, 0, 4], [0, 1, 0], [0, 0, 1]])
clockwise = rotateByDegrees(90)
print(clockwise)
oldTrap = np.array([x, y, np.ones([5])])
newTrap = np.dot(right4, oldTrap)
newTrap = np.dot(clockwise, newTrap)
# Also can use np.matmul(A,B)
# or A@B
# order matters!

# See also hstack, vstack, stack (same as vstack)
plt.figure()
plt.plot(newTrap[0, :], newTrap[1, :], lw=2)
plt.axis('square')
```
